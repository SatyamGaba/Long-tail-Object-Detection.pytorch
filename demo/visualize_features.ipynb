# Visualize feature maps

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
feature_maps = []
feature_maps.append(np.load("f_maps/initial.npy"))
for i in range(4):
    feature_maps.append(np.load(f"f_maps/{i}.npy"))
    
# print(feature_maps[0].shape)
# # feature_maps = np.array(feature_maps)
img_path = '../000000005802.jpg'
```

```python
rows, cols = 3,2
fig,ax = plt.subplots(rows, cols, figsize=(10,12))
rgb_image = Image.open(img_path)
ax[0,0].imshow(rgb_image)
chnl = 7
for i in range(1,len(feature_maps)+1):
    ax[i//cols,i%cols].imshow(feature_maps[i-1][0,chnl,...])

# plt.tight_layout()
plt.suptitle("Feature Maps at different layers")
plt.savefig("f_maps_stages.png")
plt.show()
```

```python
rows, cols = 3,2
fig,ax = plt.subplots(rows, cols, figsize=(10,12))
rgb_image = Image.open(img_path)
# ax[0,0].imshow(rgb_image)
stage = 1
for i in range(6):
    ax[i//cols,i%cols].imshow(feature_maps[stage][0,i,...])

# plt.tight_layout()
plt.suptitle("Feature Maps at different channels after one res-block")
plt.savefig("f_maps_chnls.png")
plt.show()
```
